```python
#retriving from table
from sqlalchemy import create_engine
import pandas as pd

db_engine = create_engine("postgresql+psycopg2://postgres:123456@localhost:5432/global_debt_intelligence_hub")
countries_debt = pd.read_sql("SELECT * FROM countries_debt", db_engine)
all_countries = pd.read_sql("SELECT * FROM countries", db_engine)
```

```python
len(all_countries)
```

```python
all_countries['region'].unique()
```

```python
all_countries.groupby('region')['country_code'].count().reset_index()
```

```python
#total debt by income group
total_debt_by_income_group = pd.merge(all_countries, countries_debt, on='country_code').groupby('income_group')['debt_amount'].sum().sort_values(ascending=False).reset_index()
print(total_debt_by_income_group)
```

```python
#donut chart of total debt by income group
import plotly.express as px

fig = px.pie(data_frame=total_debt_by_income_group, values='debt_amount', names='income_group', title='Total Debt by Income Group')
fig.show()
```

```python
#find  countries of high income countries
high_income_countries = all_countries[all_countries['income_group'] == 'High income']['country_name'].tolist()
print(f"High income countries: {high_income_countries}")
```

```python
#find guyana total debt 
guyana_debt = countries_debt[countries_debt['country_code'] == 'GUY']['debt_amount'].sum()
print(f"Guyana's total debt: {guyana_debt}")
```

```python
#find its percentage with total global debt
total_global_debt = countries_debt['debt_amount'].sum()
guyana_percentage = (guyana_debt / total_global_debt) * 100
print(f"Guyana's share of global debt: {guyana_percentage:.2f}%")
```

```python
#top 5 countries with highest debt
top_5_countries_debt = pd.merge(all_countries, countries_debt, on='country_code').groupby('country_name')['debt_amount'].sum().sort_values(ascending=False).head(5).reset_index()
print(top_5_countries_debt)
```

```python
# horizontal bar chart of top 5 countries with highest debt
import plotly.express as px

fig = px.bar(data_frame=top_5_countries_debt, x='debt_amount', y='country_name', orientation='h', title='Top 5 Countries with Highest Debt')
fig.show()
```

```python
#Countries that used the most  distinct indicators
top_5_countries_indicators = pd.merge(all_countries, countries_debt, on='country_code').groupby('country_name')['indicator_code'].nunique().sort_values(ascending=False).head(5).reset_index()
top_5_countries_indicators.rename(columns={'indicator_code': 'indicator_count'}, inplace=True)
print(top_5_countries_indicators)
```
