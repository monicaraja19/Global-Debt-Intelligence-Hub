```python
#retriving from table
from sqlalchemy import create_engine
import pandas as pd

db_engine = create_engine("postgresql+psycopg2://postgres:123456@localhost:5432/global_debt_intelligence_hub")

countries_debt = pd.read_sql("SELECT * FROM countries_debt", db_engine)
all_countries = pd.read_sql("SELECT * FROM countries", db_engine)
```

```python
print(len(countries_debt))
```

```python
countries_debt.describe()
```

```python
countries_debt.describe(include = object)
```

```python
region_debt =pd.merge(countries_debt,all_countries,on = 'country_code').groupby('region')["debt_amount"].mean().sort_values(ascending=False).reset_index()
region_debt
```

```python
region_debt =pd.merge(countries_debt,all_countries,on = 'country_code').groupby('region')["debt_amount"].sum().sort_values(ascending=False).reset_index()
region_debt
```

```python
import plotly.express as px

fig = px.bar(data_frame=region_debt,x="debt_amount",y="region",orientation="h",title="Total Debt by Countries Region",color="debt_amount",text="debt_amount")
```

```python
fig.show()
```

```python
income_group_debt =pd.merge(countries_debt,all_countries,on = 'country_code').groupby('income_group')["debt_amount"].sum().sort_values(ascending=False).reset_index()
income_group_debt
```

```python
fig = px.bar(data_frame=income_group_debt,x="debt_amount",y="income_group",orientation="h",title="Total Debt by Income Group",color="debt_amount",text="debt_amount")
fig.show()
```

```python
lending_category_debt =pd.merge(countries_debt,all_countries,on = 'country_code').groupby('lending_category')["debt_amount"].sum().sort_values(ascending=False).reset_index()
lending_category_debt
```

```python
fig = px.bar(data_frame=lending_category_debt,x="debt_amount",y="lending_category",orientation="h",title="Total Debt by Lending Category",color="debt_amount",text="debt_amount")
fig.show()
```

```python
external_debt_reporting_status_debt =pd.merge(countries_debt,all_countries,on = 'country_code').groupby('external_debt_reporting_status')["debt_amount"].sum().sort_values(ascending=False).reset_index()
external_debt_reporting_status_debt
```

```python
fig = px.bar(data_frame=external_debt_reporting_status_debt,x="debt_amount",y="external_debt_reporting_status",orientation="h",title="Total Debt by external_debt_reporting_status",color="debt_amount",text="debt_amount")
fig.show()
```
