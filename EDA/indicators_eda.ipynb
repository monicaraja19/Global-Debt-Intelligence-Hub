```python
#retriving from table
from sqlalchemy import create_engine
import pandas as pd

db_engine = create_engine("postgresql+psycopg2://postgres:123456@localhost:5432/global_debt_intelligence_hub")

all_indicators = pd.read_sql("SELECT * FROM indicators", db_engine)
countries_debt = pd.read_sql("SELECT * FROM countries_debt", db_engine)
```

```python
len(all_indicators)
```

```python
print(all_indicators[~all_indicators['indicator_code'].isin(countries_debt['indicator_code'])])
```

```python
#top 5 indicators with highest total debt
top_5_indicators_debt = pd.merge(all_indicators, countries_debt, on='indicator_code').groupby('indicator_code')['debt_amount'].sum().sort_values(ascending=False).head(5).reset_index()
print(top_5_indicators_debt)
```

```python
#horizontal bar chart of top 5 indicators with highest total debt
import plotly.express as px

fig = px.bar(data_frame=top_5_indicators_debt, x='debt_amount', y='indicator_code', orientation='h', title='Top 5 Indicators with Highest Total Debt')
fig.show()
```

```python
#top 5 indicator topic by total debt
top_5_indicator_topics = pd.merge(all_indicators, countries_debt, on='indicator_code').groupby('indicator_topic')['debt_amount'].sum().sort_values(ascending=False).head(5).reset_index()
print(top_5_indicator_topics)
```

```python
#horizontal bar chart of top 5 indicators topic with highest total debt
import plotly.express as px

fig = px.bar(data_frame=top_5_indicator_topics, x='debt_amount', y='indicator_topic', orientation='h', title='Top 5 Indicator Topics with Highest Total Debt')
fig.show()
```
