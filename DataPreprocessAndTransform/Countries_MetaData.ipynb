```python
import pandas as pd
from pathlib import Path

# used pathlib for getting dataset
BASE_DIR = Path.cwd().parent
DATASET_DIR = BASE_DIR / "Dataset"
#Used encoding as python default use UTF08 - but dataset contains 0x00 data which cant be loaded
countries_MetaData_df = pd.read_csv(DATASET_DIR / "Raw Data"/ "IDS_CountryMetaData.csv",encoding="latin1")
print(len(countries_MetaData_df))
```

```python
#droping completly empty rows
countries_MetaData_df.dropna(how="all",inplace=True)
print(len(countries_MetaData_df))
```

```python
#checked df dtypes
print(countries_MetaData_df.shape)
```

```python
print(countries_MetaData_df.dtypes)
```

```python
#checking for duplicates
print(countries_MetaData_df.duplicated().sum())  #no duplicates
```

```python
#checking for null/nan
print(countries_MetaData_df.isnull().sum())
```

```python
#checking if both columns have similar value
print(countries_MetaData_df[['2-alpha code','WB-2 code']][countries_MetaData_df['WB-2 code'] != countries_MetaData_df['2-alpha code']])
```

```python
print(countries_MetaData_df[['Table Name','Short Name']][countries_MetaData_df['Table Name'] != countries_MetaData_df['Short Name']])
```

```python
#droping 'WB-2 code' because 4 value differ from 2-alpha code and also its historical data we can keep 2-alpha code because its ISO standard
#droping 'Short Name' it looks like more breif,only 6 data differ from table Name 
countries_MetaData_df.drop(['WB-2 code','Short Name'],axis=1,inplace=True)
print(countries_MetaData_df.columns)
```

```python
#droping 'Other groups' has it have more that 60% of 134 data
countries_MetaData_df.drop(['Other groups','National accounts reference year','Special Notes','Vital registration complete'],axis = 1,inplace=True)
print(countries_MetaData_df.columns)
```

```python
#droping column which are least usefull
countries_MetaData_df.drop(['Long Name'],axis=1,inplace=True)
```

```python
#droping metadata related column as I focus on insights building and interpretation
countries_MetaData_df.drop(['Latest population census','Latest household survey','National accounts base year',
                                'System of National Accounts','SNA price valuation','PPP survey years',
                                'Government Accounting concept','System of trade','Balance of Payments Manual in use',
                                'Source of most recent Income and expenditure data','Latest agricultural census','Latest industrial data',
                                'Latest trade data','Latest water withdrawal data','IMF data dissemination standard'],axis=1,inplace=True)
print(countries_MetaData_df.columns)
```

```python
#Handling nan/null columns
countries_MetaData_df['External debt Reporting status'] = (countries_MetaData_df['External debt Reporting status'].fillna('Not Reported'))
countries_MetaData_df['Income Group'] = (countries_MetaData_df['Income Group'].fillna('Low income'))
print(countries_MetaData_df.columns)
```

```python
#checking for null/nan
print(countries_MetaData_df.isnull().sum())
```

```python
#removing aggregated countries
countries_MetaData_df.dropna(subset=['Region'],inplace=True)
```

```python
#checking for null/nan
print(countries_MetaData_df.isnull().sum())
```

```python
print(len(countries_MetaData_df))
```

```python
#Standardizing the order of column
countries_MetaData_df = countries_MetaData_df[['Code','Table Name','2-alpha code','Currency Unit','Region','Income Group','Lending category','External debt Reporting status']]
print(countries_MetaData_df.head(2))
```

```python
#standardzing primary column
countries_MetaData_df['Code'] =countries_MetaData_df['Code'].str.strip()
```

```python
print(countries_MetaData_df['Code'].duplicated().sum())
```

```python
#renaming column as that of table name
countries_MetaData_df = countries_MetaData_df.rename(columns={'Code': 'country_code','Table Name': 'country_name','2-alpha code': 'alpha2_code',
                                                              'Currency Unit':'currency_unit','Region':'region','Income Group':'income_group',
                                                              'Lending category':'lending_category','External debt Reporting status':'external_debt_reporting_status'})
```

```python
countries_MetaData_df.to_csv(DATASET_DIR/ "Processed Data" / "CountryMetaData.csv",index=False)
```
