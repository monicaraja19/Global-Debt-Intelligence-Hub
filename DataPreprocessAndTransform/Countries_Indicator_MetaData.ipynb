```python
import pandas as pd
from pathlib import Path

# used pathlib for getting dataset
BASE_DIR = Path.cwd().parent
DATASET_DIR = BASE_DIR / "Dataset"
#Used encoding as python default use UTF08 - but dataset contains 0x00 data which cant be loaded
countries_Indicator_MetaData_df = pd.read_csv(DATASET_DIR / "Raw Data"/ "Country-Series - Metadata.csv",encoding="latin1")
print(len(countries_Indicator_MetaData_df))
```

```python
#droping completly empty rows
countries_Indicator_MetaData_df.dropna(how="all",inplace=True)
print(len(countries_Indicator_MetaData_df))
```

```python
#checked df dtypes
print(countries_Indicator_MetaData_df.shape)
print(countries_Indicator_MetaData_df.dtypes)
```

```python
#checking for duplicates
print(countries_Indicator_MetaData_df.duplicated().sum())  #no duplicates
```

```python
#checking for null/nan
print(countries_Indicator_MetaData_df.isnull().sum())
```

```python
print(countries_Indicator_MetaData_df.head(2))
```

```python
print(countries_Indicator_MetaData_df['Type'].unique())
```

```python
#Droping column that is not usefull as it has same value
countries_Indicator_MetaData_df.drop(['Type'],axis=1,inplace=True)
print(countries_Indicator_MetaData_df.isnull().sum())
```

```python
print(countries_Indicator_MetaData_df.head(2))
```

```python
#Standardized some of the columns for relationship with other table/dataset
countries_Indicator_MetaData_df['CountryCode'] = countries_Indicator_MetaData_df['Country Code'].str.split('(',regex=False).str[1].str.replace(')', '', regex=False)
countries_Indicator_MetaData_df['SeriesCode'] = countries_Indicator_MetaData_df['Series Code'].str.split('(',regex=False).str[-1].str.replace(')', '', regex=False)
print(countries_Indicator_MetaData_df.head(2))
```

```python
#Standardizing the order of column
countries_Indicator_MetaData_df = countries_Indicator_MetaData_df[['CountryCode','SeriesCode','Description']]
print(countries_Indicator_MetaData_df.head(2))
```

```python
print(len(countries_Indicator_MetaData_df))
```

```python
print(countries_Indicator_MetaData_df[countries_Indicator_MetaData_df[['CountryCode', 'SeriesCode']].duplicated()])
```

```python
print(countries_Indicator_MetaData_df[countries_Indicator_MetaData_df[['CountryCode', 'SeriesCode','Description']].duplicated()])
```

```python
#droping duplictaes
countries_Indicator_MetaData_df.drop_duplicates(inplace=True)
```

```python
print(countries_Indicator_MetaData_df[['CountryCode', 'SeriesCode']].value_counts())
```

```python
print(len(countries_Indicator_MetaData_df))
```

```python
#renaming column as that of table name
countries_Indicator_MetaData_df = countries_Indicator_MetaData_df.rename(columns={'CountryCode': 'country_code','SeriesCode': 'indicator_code','Description': 'description'})
```

```python
#standardizing the primary columns
countries_Indicator_MetaData_df['country_code'] = countries_Indicator_MetaData_df['country_code'].str.strip()
countries_Indicator_MetaData_df['indicator_code'] = countries_Indicator_MetaData_df['indicator_code'].str.strip()
```

```python
print(countries_Indicator_MetaData_df[countries_Indicator_MetaData_df["country_code"].str.len() != 3])
```

```python
countries_Indicator_MetaData_df.drop(92,inplace=True)
```

```python
#checking for valid primary/foreign column values
countries_df = pd.read_csv(DATASET_DIR / "Processed Data"/ "CountryMetaData.csv",encoding="latin1")
valid_countries = countries_df["country_code"]
print(valid_countries.nunique())
```

```python
#filtering out invalid countries
invalid_codes = countries_Indicator_MetaData_df[~countries_Indicator_MetaData_df["country_code"].isin(valid_countries)]
print(invalid_codes["country_code"].unique())
```

```python
#removing invalid country rows
countries_Indicator_MetaData_df = countries_Indicator_MetaData_df[countries_Indicator_MetaData_df["country_code"].isin(countries_df["country_code"])]
print(len(countries_Indicator_MetaData_df))
```

```python
countries_Indicator_MetaData_df.to_csv(DATASET_DIR/ "Processed Data" / "Country_Series_Metadata.csv",index=False)
```
